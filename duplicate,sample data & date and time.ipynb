```python
import pandas as pd
employee={"Name":["Alex","Max","Mira","Ayesha","Pooja"],
          "Salary":[30000,50000,40000,55000,45000]}
eply=pd.DataFrame(employee)
print(eply)
```

```python
print(eply[eply["Salary"]>45000])   #manual way
```

```python
print(eply.query("Salary>45000"))  #query method -use for filtering
```

```python
print(eply.query("Salary<55000 and Salary>40000"))
```

#iteration 

```python
fruits=["mango","orange","apple","banana"]
it=iter(fruits)
print(next(it))
print(next(it))
print(next(it))
```

```python
studentmarks=[122,133,144,124]
s=iter(studentmarks)
print(next(s))
print(next(s))
print(next(s))
```

```python
def numbers():
    yield 13
    yield 2
    yield 3
gen = numbers()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
try:
    num=int(input("Enter the number:"))
    print(10/num)
except ZeroDivisionError:
    print("number is not zero")
except ValueError:
    print("enter only numerical digits")
except TypeError:
    print("enter integer")
```

```python
try:
    num=int(input("Enter the number"))
    print(10/num)
except ValueError:
    print("Invalid Inout")
except ZeroDivisionError:
    print("Cannot divide by Zero")
else:
    print("program execute sucessfully")
finally:
    print("This block is going to execute everytime")
```

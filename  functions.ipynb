```python
def add(a,b,c=0):
    return a+b+c
print(add(11,12))
print(add(11,22,14))
```

```python
def mul(a,b,c=0):
    return a*b*c
print(mul(12,13))
print(mul(12,13,15))
```

```python
def add(*numbers):
    print(sum(numbers))
add(10,20)
add(10,20,30)
add(10,20,30,40)
add(30,40,50,100,40)
```

```python
def students(*marks):
    print(sum(marks))
students(30,40,50,6)
students(30,33,17,30,47)
students(14,25,16,19,20,35)
```

```python
s=int(input("Enter the number:"))
def factorial(x):
    if x==1:
        return 1
    else:
        return x*factorial(x-1)
print(factorial(x))   
        
```

```python
def prime(s):
    if s<=1:
        return
    else:
        for i in range(2,s):
            if s%i==0:
                return ("not prime")
        else:      
          return("prime")
print(prime(2))        
        
```

```python
s="hello"
vow="aeiouAEIOU"
def vowels():
 count=0
 for char in s:
    if char in vow:
        count +=1
 return count    
         
print("vowels are:",vowels())    
```

#OOPs

```python
class Student:
    def __init__(self,name):
        self.name = name
    def display(self):
        print("Name:",self.name)
s1=Student("Abdul")
s1.display()
        
```

```python
class Employee:
    id="12345"
    name="Meera"
    salary="50000"
emp1=Employee
print(emp1.id,emp1.name,emp1.salary)
    
```

```python
class Car:
    brand="BMW"
    model="2021"
car=Car
print(car.brand)
print(car.model)
    
```

```python
class StudentGrade:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
s1 = StudentGrade("Rani",87)
print(s1.name)
print(s1.marks)
        
```

```python
class Price:
    book=11
    pen=5
    def display(self):
        print(self.book,self.pen)
p1=Price()
p1.display()
```

#inheritance

```python
class Person:
    def raw(self):
        print("tall")
class Student(Person):
    def raw1(self):
        print("intelligent")
obj=Student()
obj.raw()
obj.raw1()
        
        
```

```python
class Vehicle:
    colour="black"
    model="2022"
class Car(Vehicle):
    brand="mercedes"
c1=Car
print(c1.colour+" "+c1.model+" "+c1.brand)
    
    
    
```

```python
class Parent:
   def hair(self):
    print("hair is black in colour")
class Child(Parent):
   def eye(self):
    print("eye colour is blue")
object=Child()
object.hair()
object.eye()
    
    
```

```python
class A:
    def func1(x):
        print("tall")
class B(A):
    def func2(x):
        print("thin")
class C(B):
    def func3(x):
        print("fair")
object=C()
object.func1()
object.func2()
object.func3()
        
        
```

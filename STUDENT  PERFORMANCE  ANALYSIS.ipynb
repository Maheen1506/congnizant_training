```python
import pandas as pd

file_path = r"C:\Users\dell\Documents\student.csv.xlsx"
df = pd.read_excel(file_path)

print(df)
```

```python
df.isnull()     #HANDLING MISSING VALUES
```

```python
df.fillna(76,inplace = True)
```

```python
total=df["Percentage"].sum()         #TOTAL
```

```python
print(total)
```

```python
Average=[total/20]
print(Average)                    #AVERAGE
```

```python
df["Percentage"].mean()      #MEAN
```

```python
df["Percentage"].median()    #MEDIAN
```

```python
df["Percentage"].mode()      #MODE
```

```python
def grade(avg):
    if avg>=90:
        return'A'
    elif avg>=75:                  #GRADE
        return'B'
    elif avg>=60:
        return'C'
    else:
        return'D'
df["Grade"]=df["Percentage"].apply(grade)

print("\nstudent results")
print(df)
        
        
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.bar(df['Name'],df['Percentage'])
plt.xlabel('student names')                   #BARGRAPH
plt.ylabel('Percentage')
plt.title("Percentage Of Students Marks")
plt.show()
```

```python
fig=plt.figure(figsize=(7,7))

plt.pie(df['Percentage'],labels = df["Subject"])

plt.legend(                                              #PIECHART
            title="Subject",
    loc= 'lower right')
plt.title("SUBJECTS PIE CHART")
plt.show()
```

```python
plt.hist(df['Grade'],bins=3,color="Blue",edgecolor='black')
plt.xlabel("Grade")
plt.ylabel("frequency")                   #HISTOGRAM
plt.title("Grade Distributions")
plt.show()
```

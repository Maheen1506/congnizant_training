```python
#SNOWBALL SAMPLING(reference base selection)
networks={"n1":["n2","n3"],
          "n2":["n4"],
          "n3":["n5"]}
sample=["n1"]
sample.extend(networks["n1"])
sample.extend(networks["n2"])
print("SnowBall sampling:",sample)
```

```python
import numpy as np
product=["electronics","clothes","Jwellery"]
jansales=[200,300,250]
febsales=[300,250,210]
marchsales=[250,200,250]
x=np.arange(len(product))
width=0.5
plt.bar(x-width/3,jansales,width,label="jan")
plt.bar(x+width/3,febsales,width,label="feb")
plt.bar(x+width/2,marchsales,width,label="march")
plt.xlabel("product")
plt.title("sales comparision:jan vs feb vs vs march")
plt.xticks(x,product)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
product=["Laptop","Mobile","bluetooth"]
online=[200,300,350]
offline=[250,450,340]
wholesale=[300,480,390]
plt.bar(product,online,label='online')
plt.bar(product,offline,bottom=online,label='offline')

bottom_value=[online[i]+offline[i] for i in range(len(product))]
plt.bar(product,wholesale,bottom=bottom_value,label="wholesale")
plt.xlabel("product")
plt.ylabel("revenue")
plt.title("Revenue of the Distributed sales")
plt.xticks(x,product)
plt.legend()
plt.show()
```

```python
num=135
num=abs(num)
count=0 
if num==0:
 count=1
else:
 while num>0:
    count+=1
    num//=10
print("The Number of Digits:",count)    
```

```python
s=123
reversed_number=0
while s!=0:
    digit=s%10
    reversed_number=(reversed_number*10)+digit
    s=s//10
print("Reversed Number:",reversed_number)
    
```

```python
num=int(input("Enter the Number:"))
n=num
reversed_num=0
while n>0:
    digit=n%10
    reversed_num=(reversed_num*10)+digit
    n=n//10
if num==reversed_num:
    print("the digit is palindrome")
else:
    print("the digit is not a palindrome")
```

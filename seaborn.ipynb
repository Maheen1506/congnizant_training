```python
import seaborn as sns
import matplotlib.pyplot as plt
months=["Jan","Feb","March","April","May"]
sales=[200,300,400,350,250]
sns.lineplot(x=months,y=sales)
plt.xlabel("months")
plt.ylabel("sales")
plt.title("Monthly Revenue")
plt.show()
```

```python
import plotly.express as px
import pandas as pd
data=pd.DataFrame({"Months":["May","June","July","Aug"],
      "Profit":[300,400,500,600]})
graph=px.line(data,x="Months",y="Profit",markers=True)
graph.show()


```

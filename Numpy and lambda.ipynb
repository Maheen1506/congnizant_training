```python
def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
numbers=[1,2,3,4,5]
result=list(map(lambda E:E*E,numbers))
print(result)
```

```python
#filter
```

```python
num=[13,12,14,15]
result=list(filter(lambda M:M%2==0,num))
print(result)
```

```python
num=[13,12,14,15]
result=list(filter(lambda M:M%2!=0,num))
print(result)
```

```python
from functools import reduce
num=[13,12,14,15]
result=reduce(lambda X,Y:X+Y,num)
print(result)
```

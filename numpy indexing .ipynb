#iteration

```python
marks=np.array([30,49,50,60,80])
for i in marks:
    print(i)
```

```python
l=[13,14,15,16]
for i in range(13,17):
    print(i)
```

```python
import numpy as np
s=np.array([[12,13,14,15],[14,25,36,75]])
for row in s:
        print(row)
```

```python
for i in np.nditer(s):
    print(i)
```

```python
e=np.array([[40,50,50],[40,60,70],[40,60,80]])
for i in np.nditer(e):
    print(i)
```

#merging or joining two array

```python
w=np.array([40,30,45])
s=np.array([35,65,70])
c=np.concatenate((w,s))
print(c)
```

```python
w=np.array([40,30,45])
s=np.array([35,65,70])
c=(w+s)
print(c)
```

#mathematical operations

```python
marks=np.array([45,56,78,90,89])
print(np.sum(marks))
print(np.average(marks))
print(np.mean(marks))
print(np.max(marks))
print(np.min(marks))
print(np.square(marks))
print(np.power(marks,3))
```

#arrange

```python
arr=np.arange(10)
print(arr)
```

```python
arr=np.arange(10)
print(arr[::2])        #skip every 2nd element
```

```python
print(arr[1:7:2])    #from index 1 to 6 and skipping every 2nd element
```

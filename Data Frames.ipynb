```python
students=pd.DataFrame({
    "ID":[12,13,14,15],
    "Name":["Max","Rahul","Abdul","Rimsha"]})
print(students)
```

```python
students2=pd.DataFrame({
    "ID":[12,13,14,15],                       
    "Marks":[70,80,90,78]})
print(students2)
```

```python
students=pd.DataFrame({
    "ID":[12,13,14,15],
    "Name":["Max","Rahul","Abdul","Rimsha"]})
students2=pd.DataFrame({
    "ID":[12,13,14,15],                       #MERGING OF TWO DATA TABLES
    "Marks":[70,80,90,78]})
result=pd.merge(students,students2,on="ID")
print(result)
                      
```

```python
pd.merge(students,students2,how="inner",on="ID")
```

```python
students=pd.DataFrame({
    "ID":[12,13,14,15],
    "Name":["Max","Rahul","Abdul","Rimsha"]})
students2=pd.DataFrame({
    "ID":[12,13,14,15],                       #MERGING OF TWO DATA TABLES
    "Marks":[70,80,90,78]})
result=pd.merge(students,students2,on="ID")
print(result)
                      
```

```python
pd.merge(students,students2,how="right",on="ID")
```

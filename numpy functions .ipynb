```python
import numpy as np
arr=np.array([10,20,30,40,50,60])
print(arr)
print(arr[arr>50])
print(arr[[1,2]])
print(np.sum(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.max(arr))
print(np.min(arr))
print(arr*3)
print(np.sort(arr))
print(np.hstack(arr))
print(np.split(arr,2))
print(np.where(arr==50))
print(np.power(arr,3))
for i in arr:
    print(i)
for i in np.nditer(arr):
    print(i)
    
print(np.delete(arr,1))      
```

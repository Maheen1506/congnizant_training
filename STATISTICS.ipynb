```python
import numpy as np
salary=([50000,60000,70000,89000])
print(np.mean(salary))
print(np.median(salary))
print(np.max(salary))
print(np.min(salary))
print(np.max(salary)-np.min(salary))
print(np.var(salary))
print(np.std(salary))

```

#Statstical Operations

```python
import pandas as pd
w=pd.Series([10,20,30,40,50,60])
print(w.sum())
print(w.max())
print(w.min())
print(w.mean())
print(w.std())
print(w.var())

```

#Boolean Filtering

```python
import pandas as pd
w=pd.Series([10,20,30,40,50,60])
print(w[w>=30])
print([w>=20])

```

```python
print(w==20)
```

#Missing value in Series

```python
import pandas as pd
import numpy as np
s=pd.Series([10,20,30,np.nan,50])
print(s)
print(s.isnull())
print(s.isnull().sum())#to count no of empty space
print(s.fillna(2))
print(s.dropna())
```

```python
s=pd.Series(["mars","venus","neptune","saturn","uranus"])
print(s)
print(s.str.upper())          #UPPER AND LOWER CASE 
print(s.str.lower())
```

```python
q=pd.Series([12,23,45,24,46,13])  #SORTING OF ARRAY
print(q.sort_values())           #SORTING ACCORDING TO DESCENDING ORDER
print(q.sort_values(ascending=False))
```

```python
e=pd.Series([20,40,30,28,20,30,45,17,28])#TO GET UNIQUE ELEMENTS
print(e.unique())
```

```python
e=pd.Series([20,40,30,28,20,30,45,17,28])   #TO COUNT EACH VALUE IN SERIES ARRAY
print(e.value_counts())                       
```

```python
e=pd.Series([20,40,30,28,20,30,45,17,28])     #CONVERTING SERIES TO LIST
print(e.tolist())
```
